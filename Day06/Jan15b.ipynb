```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
data = load_boston()
dir(data)
```

```python
data["DESCR"]
```

```python
data["data"]
```

#### Creating Dataframe

```python
dfX = pd.DataFrame(data["data"], columns = data["feature_names"])
dfX.head()
```

```python
dfY = pd.DataFrame(data["target"], columns= ["price"])
dfY
```

```python
ppS = preprocessing.scale(dfX)
ppS
```

```python
dfXs = pd.DataFrame(ppS, columns= data["feature_names"])
dfXs["price"] = dfY
dfXs
```

```python
dfXs.corr()
```

```python
plt.figure(figsize= (12,8))
sb.heatmap(dfXs.corr(), annot= True)
plt.show()
```

Dari korelasi terlihat data yang berpengaruh ke harga adalah RM, PTRATIO, dan LSTAT

#### Splitting

```python
x = dfXs[["RM", "PTRATIO", "LSTAT"]]
y = dfXs["price"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.1)
```

```python
print(x_train.iloc[0])
print(y_train[x_train.iloc[0].name])
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

#### Membandingkan harga prediksi dan aktual

```python
print(model.predict([x_test.iloc[0]]))
print(y_test[x_test.iloc[0].name])
```

#### Error checking

```python
hargaPred = model.predict(x_test)
harga = y_test
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(harga, hargaPred)
rmse = np.sqrt(mse)
rmse
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(harga, hargaPred)
r2
```

```python
model.score(x_test, y_test)
```
