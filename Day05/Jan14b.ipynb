```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_excel("Normalisasi.xlsx", header= 1)
df = df["TB (m)"].iloc[:10]
df.name = "TB"
df = pd.DataFrame(df)
df
```

#### Import Method

```python
from sklearn.preprocessing import StandardScaler as SSc
from sklearn.preprocessing import MinMaxScaler as MMSc
from sklearn.preprocessing import MaxAbsScaler as MASc
from sklearn.preprocessing import RobustScaler as Rosc
from sklearn.preprocessing import Normalizer as Nor
from sklearn.preprocessing import QuantileTransformer as Qtr
from sklearn.preprocessing import Binarizer as Bnr
```

#### Normalisasi Z-score

```python
df["Z Std"] = SSc().fit_transform(df[["TB"]])
df
```

#### Normalisasi Min Max Scaler

```python
df["MinMax"] = MMSc().fit_transform(df[["TB"]])
df
```

#### Max Abs Scaler

```python
df["MaxAbs"] = MASc().fit_transform(df[["TB"]])
df
```

#### Robust Scaler

```python
df["Robust"] = Rosc().fit_transform(df[["TB"]])
df
```

#### Quantile Transformer/Gaussian

```python
df["Quantile1"] = Qtr(output_distribution = "normal").fit_transform(df[["TB"]])
df
```

```python
df["Quantile2"] = Qtr(output_distribution = "uniform").fit_transform(df[["TB"]])
df
```

#### Normalizer

```python
df["Norm"] = Nor().fit_transform(df[["TB"]])
df
```

#### Binarizer

```python
df["Biner"] = Bnr(threshold = 0.85).fit_transform(df[["TB"]])
df
```
