```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
from sklearn.datasets import load_breast_cancer
bc = load_breast_cancer()
dir(bc)
```

```python
df = pd.DataFrame(bc["data"], columns= bc["feature_names"])
df["target"] = bc["target"]
df.head()
```

```python
df.isnull().sum()
```

```python
df.corr()["target"]
```

#### Cek yang korelasinya lebih dari 0.5 atau kurang dari -0.5

```python
df.corr()["target"][df.corr()["target"] >.5]
```

```python
df.corr()["target"][df.corr()["target"] <-.5]
```

```python
i = np.array(df.corr()["target"][df.corr()["target"] <-.5].index)
dfNew = df[i]
dfNew["target"] = df["target"]
dfNew
```

### Standard Scalling

```python
from sklearn.preprocessing import StandardScaler
```

```python
stds = StandardScaler()
dataSTD = stds.fit_transform(dfNew.iloc[:,:-1]) #exclude target
dataSTD
```

#### Cek & Drop Outliers

```python
dfZ = pd.DataFrame(dataSTD)
dfZ["target"] = dfNew["target"]
dfZ
```

```python
dfZ[(dfZ > 2.5) | (dfZ < -2.5)] = np.NaN
dfZ = dfZ.dropna()
dfZ
```

```python
dataSTD2 = np.array(dfZ.iloc[:,:-1])
dataSTD2
```

### PCA

```python
from sklearn import decomposition
```

```python
pca = decomposition.PCA(n_components= 3)

datapca = pca.fit_transform(dataSTD2)
datapca
```

```python
pca.explained_variance_ratio_
```

### Pipeline : StandardScaller & PCA

```python
from sklearn.pipeline import make_pipeline
pipe = make_pipeline(StandardScaler(), decomposition.PCA(n_components= 3))
datapipe = pipe.fit_transform(dfNew.iloc[:,:-1])
datapipe
```

#### Inverse ke value semula

```python
pipe.inverse_transform(datapipe)[0]
```

### Splitting Datasets

```python
x = datapca
y = np.array(dfZ["target"])

x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
xtr, xts, ytr, yts = train_test_split(x, y, test_size= .15)

xtr.shape, xts.shape
```

### K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
k = round(xtr.shape[0] ** .5)
k
```

```python
modKNN = KNeighborsClassifier(n_neighbors= k)
modKNN.fit(xtr, ytr)
```

```python
from sklearn.metrics import classification_report
print(classification_report(yts, modKNN.predict(xts)))
```

```python
modKNN.score(xts, yts)
```
