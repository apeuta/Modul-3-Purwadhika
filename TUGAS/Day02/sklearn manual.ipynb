```python
import numpy as np
import pandas as pd
```

```python
data = {
    'mesin': [1000, 2000, 3000, 4000, 5000],
    'harga': [10, 25, 35, 55, 80]
}
df = pd.DataFrame(data)
df
```

```python
x2 = []
for i in df['mesin']:
    x2.append(i*i)
df['x^2'] = x2

y2 = []
for i in df['harga']:
    y2.append(i*i)
df['y^2'] = y2

xy = []
for i in range(len(df['mesin'])):
    xy.append(df['mesin'].iloc[i]*df['harga'].iloc[i])
df['xy'] = xy

df
```

```python
slope = ((len(df['mesin'])*(df['xy'].sum()))-(df['mesin'].sum()*(df['harga'].sum())))/((len(df['mesin'])*(df['x^2'].sum()))-(np.square(df['mesin'].sum())))
slope
```

```python
incept = (((df['harga'].sum())*(df['x^2'].sum()))-(df['mesin'].sum()*(df['xy'].sum())))/((len(df['mesin'])*(df['x^2'].sum())) - np.square(df['mesin'].sum()))
incept
```

```python
yPred = []
for i in df['mesin']:
    yPred.append(slope * i + incept)
df["y'"] = yPred
df
```

```python
selisihY = []
for i in range(len(df['harga'])):
    selisihY.append(np.abs(df['harga'].iloc[i] - df["y'"].iloc[i]))
df["|y - y'|"] = selisihY
df
```

```python
kdrtSelisihY = []
for i in df["|y - y'|"]:
    kdrtSelisihY.append(np.square(i))
df["|y - y'|^2"] = kdrtSelisihY
df
```

```python
selisihRerata = []
for i in range(len(df['harga'])):
    selisihRerata.append(np.abs(df['harga'].iloc[i] - np.mean(df['harga'])))
df['|y - avg(y)|'] = selisihRerata
df
```

```python
sqrSelisih = []
for i in df['|y - avg(y)|']:
    sqrSelisih.append(np.square(i))
df['|y - avg(y)|^2'] = sqrSelisih
df
```

```python
eLog = []
for i in df['harga']:
    eLog.append(np.log(1 + i))
df['elog(1 + y)'] = eLog
df
```

```python
eLogAks = []
for i in df["y'"]:
    eLogAks.append(np.log(1 + i))
df["elog(1 + y')"] = eLogAks
df
```

```python
eLogVal = []
for i in range(len(df['elog(1 + y)'])):
    eLogVal.append(np.square(df['elog(1 + y)'].iloc[i] - df["elog(1 + y')"].iloc[i]))
df["(elog(1 + y) - elog(1 + y'))^2"] = eLogVal
df
```

```python
pctError = []
for i in range(len(df["|y - y'|"])):
    pctError.append(np.abs(df["|y - y'|"].iloc[i]/df['harga'].iloc[i]))
df["%error"] = pctError
df
```

```python
maxOfError = np.max(df["|y - y'|"])
meanAbsoluteError = (df["|y - y'|"].sum())/len(df["mesin"])
meanSquaredError = np.square(meanAbsoluteError)
squareRootMSE = np.sqrt(meanSquaredError)
medianAbsoluteError = np.median(df["|y - y'|"])
meanSquaredLogarithmicError = (df["(elog(1 + y) - elog(1 + y'))^2"].sum())/len(df["(elog(1 + y) - elog(1 + y'))^2"])
squaredMSLE = np.sqrt(meanSquaredLogarithmicError)
r2Score = 1-(df["|y - y'|^2"].sum())/(df["|y - avg(y)|^2"].sum())
meanAbsolutePercentageError = df['%error'].sum()/len(df['%error'])
print("Max Of Error: ", round(maxOfError, 3))
print("Mean Absolute Error: ", round(meanAbsoluteError, 3))
print("Mean Squared Error: ", round(meanSquaredError, 3))
print("Root Squared Mean Squared Error: ", round(meanSquaredError, 3))
print("Median Absolute Error: ", round(medianAbsoluteError, 3))
print("Mean Squared Logarithmic Error: ", round(meanSquaredLogarithmicError, 3))
print("Square MSLE: ", round(squaredMSLE, 3))
print("R2 Score: ", round(r2Score, 3))
print("Mean Absolute Percentage Error: ", round(meanAbsolutePercentageError, 3))
```
