```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sb
```

```python
df = pd.read_csv("insurance.csv")
df
```

```python
df.corr()
```

```python
ududs = []
for i in df["smoker"]:
    if i == "yes":
        ududs.append(1)
    else:
        ududs.append(0)
df["ududs"] = ududs

jKel = []
for i in df["sex"]:
    if i == "female":
        jKel.append(1)
    else:
        jKel.append(0)
df["jKel"] = jKel
df
```

```python
df.corr()
```

```python
model = LinearRegression()
model.fit(df[["age", "bmi", "ududs"]], df["charges"])
df["Best Price"] = model.predict(df[["age", "bmi", "ududs"]])
df
```

```python
sb.heatmap(df.corr(),annot=True, cmap= "hot_r")
plt.show()
```
